## Sum function

```python
def sum(inputs, weights):
  s = 0
  for i in range(2):
    s += inputs[i] * weights[i]
  return s
```

```python
s = sum(inputs, weights)
```

```python
s
```

## Step function

```python
def step_function(sum):
  if (sum >= 1):
    return 1
  return 0
```

## Final result

```python
step_function(s)
```
