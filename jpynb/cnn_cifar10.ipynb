# Small Image Classification Using Convolutional Neural Network (CNN)

In this notebook, we will classify small images from the CIFAR-10 dataset using TensorFlow/Keras.

**CIFAR-10 Dataset contains 10 classes:**
- airplane, automobile, bird, cat, deer, dog, frog, horse, ship, truck

**We will compare:**
1. Artificial Neural Network (ANN) - baseline
2. Convolutional Neural Network (CNN) - improved accuracy

![CIFAR-10 Classes](../images/small_images.jpg)

````python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np

print(f"TensorFlow version: {tf.__version__}")
````

## Load the Dataset

CIFAR-10 contains:
- 50,000 training images
- 10,000 test images
- 10 classes
- Image size: 32x32x3 (RGB)

````python
(X_train, y_train), (X_test, y_test) = datasets.cifar10.load_data()

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
print(f"Training labels shape: {y_train.shape}")
````

````python
# y_train is a 2D array, convert to 1D for classification
y_train = y_train.reshape(-1,)
y_test = y_test.reshape(-1,)

print(f"y_train shape after reshape: {y_train.shape}")
print(f"First 5 labels: {y_train[:5]}")
````

````python
# Define class names
classes = ["airplane", "automobile", "bird", "cat", "deer", 
           "dog", "frog", "horse", "ship", "truck"]
print(f"Number of classes: {len(classes)}")
````

## Visualize Sample Images

Let's plot some images to see what they look like.

````python
def plot_sample(X, y, index):
    """Plot a single sample image with its label."""
    plt.figure(figsize=(15, 2))
    plt.imshow(X[index])
    plt.xlabel(classes[y[index]])
    plt.title(f"Class: {classes[y[index]]}")
    plt.show()
````

````python
# Plot first two samples
plot_sample(X_train, y_train, 0)
plot_sample(X_train, y_train, 1)
````

## Normalizing the Training Data

Normalize images to a range from 0 to 1. Each RGB channel value ranges from 0 to 255, so we divide by 255.

````python
X_train = X_train / 255.0
X_test = X_test / 255.0
print("Data normalized to range [0, 1]")
````

## Build Simple Artificial Neural Network (ANN) for Image Classification

First, let's try a simple ANN as a baseline to compare against CNN.

````python
ann = models.Sequential([
    layers.Flatten(input_shape=(32, 32, 3)),
    layers.Dense(3000, activation='relu'),
    layers.Dense(1000, activation='relu'),
    layers.Dense(10, activation='softmax')    
])

ann.compile(optimizer='SGD',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])

ann.fit(X_train, y_train, epochs=5)
````

**You can see that at the end of 5 epochs, accuracy is at around 49%**

````python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = ann.predict(X_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("ANN Classification Report:")
print(classification_report(y_test, y_pred_classes, target_names=classes))
````

## Build a Convolutional Neural Network (CNN)

Now let's build a CNN to train our images. CNN uses:
- **Conv2D layers** - to extract features from images
- **MaxPooling2D layers** - to reduce dimensionality while preserving important features
- **Dense layers** - for final classification

````python
cnn = models.Sequential([
    # First Convolutional Block
    layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    
    # Second Convolutional Block
    layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    # Third Convolutional Block
    layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    # Dense Layers
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
````

````python
cnn.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])

cnn.summary()
````

````python
# Train the CNN
cnn.fit(X_train, y_train, epochs=10)
````

**With CNN, accuracy reaches around 70% which is a significant improvement over ANN!**

CNNs are best for image classification because:
- MaxPooling reduces image dimensions while preserving features
- Conv2D layers learn spatial hierarchies of features
- Less computation compared to simple ANN

````python
# Evaluate CNN on test data
cnn.evaluate(X_test, y_test)
````

````python
# Make predictions
y_pred = cnn.predict(X_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

# Show first 5 predictions
print("First 5 predictions:", y_pred_classes[:5])
print("First 5 actual:", y_test[:5].tolist())
````

````python
# Visualize a test prediction
test_idx = 3
plot_sample(X_test, y_test, test_idx)
print(f"Predicted class: {classes[y_pred_classes[test_idx]]}")
````

````python
# CNN Classification Report
print("CNN Classification Report:")
print(classification_report(y_test, y_pred_classes, target_names=classes))
````
